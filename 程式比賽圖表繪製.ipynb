#### 將上述數據的平均繪製成密度分布圖呈現

```python
#需使用到表格中全台縣市的經緯度
nep_districts = gpd.read_file(r'TWN_adm/TWN_adm2.shp')#使用geopandas讀取
nep_districts
```

```python
#取出縣市名稱及經緯度
nep_districts = nep_districts[['NAME_2', 'geometry']]
nep_districts.rename(columns = {'NAME_2' : 'District'}, inplace = True)#將Name_2重新命名為District
nep_districts
```

```python
#全台火災發生率平均數據
data4 = pd.read_csv("發生率平均.csv")
data4
```

```python
#全台火災發生次數平均數據
data5 = pd.read_csv("發生次數平均.csv")
data5
```

```python
#全台火災死傷率平均數據
data6=pd.read_csv("死傷率平均.csv")
data6
```

```python
#將全台縣市經緯度的英文翻譯與數據統一
nep_districts.replace('Taipei City', 'New Taipei', inplace = True)
nep_districts.replace('Changhwa', 'Changhua', inplace = True)
nep_districts.replace('Ilan', 'Yilan', inplace = True)
nep_districts.replace('Keelung City', 'Keelung', inplace = True)
nep_districts
```

```python
#表格合併:全台縣市經緯度&全台火災發生率平均
a=nep_districts.merge(data4, on = 'District')
a
```

```python
#使用geopandas匯出密度分布圖
a.plot(column = 'Occurance Rate Average', cmap = 'coolwarm', legend = True)
```

```python
#表格合併:全台縣市經緯度&全台火災發生次數平均
b=nep_districts.merge(data5, on = 'District')
b
```

```python
#使用geopandas匯出密度分布圖
b.plot(column = 'Occurance Average', cmap = 'coolwarm', legend = True)
```

```python
#表格合併:全台縣市經緯度&全台火災死傷率平均
c=nep_districts.merge(data6, on = 'District')
c
```

```python
#使用geopandas匯出密度分布圖
c.plot(column = 'Mortality Rate Average', cmap = 'bwr', legend = True)
```

#### 全台火災發生之季節比以圓餅圖呈現

```python
#全台發生火災季節比
labels = 'Spring','Summer','Autumn','Winter'
size = [31.7,16.5,23,28.8]#設定每一區塊比例
colors = ['salmon','lightsteelblue','steelblue','darkgray']
separated = (.05,0,0,0)#將比例最高的春季分離出來
plt.pie(size , labels = labels,colors=colors,autopct='%1.1f%%',explode=separated)#將上述的數值帶入
plt.axis('equal')#讓每個區塊相等，方便帶入數值
plt.show()
```
